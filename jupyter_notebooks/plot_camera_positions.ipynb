```python
calibrated_cameras = {
    '1': camera_1[0:3, 3],
    '2': camera_2[0:3, 3],
    '3': camera_3[0:3, 3],
    '4': camera_4[0:3, 3],
    '5': camera_5[0:3, 3],
    '6': camera_6[0:3, 3],
    '7': camera_7[0:3, 3],
    '8': camera_8[0:3, 3],
    '12': camera_12[0:3, 3],
    '13': camera_13[0:3, 3]
}
```

```python
# Calculate errors
errors = []

for camera in calibrated_cameras.keys():
    errors.append(abs(ground_truth_cameras[camera] - calibrated_cameras[camera]))
    
errors = np.array(errors)
print(f'Mean Error: {errors.mean(axis=0)}')
print(f'Median Error: {np.median(errors, axis=0)}')
print(f'Std. Deviation {np.std(errors, axis=0)}')
# print(f'Variance {np.var(errors, axis=0)}')
```
